```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def gme_revenue(url: str):
    # Fetch the HTML content from the URL
    data = requests.get(url).text
    
    # Parse the HTML content
    soup = BeautifulSoup(data, 'html.parser')
    
    # Initialize an empty list to store data
    revenue_data = []

    # Extract data from the table
    for row in soup.find("tbody").find_all('tr'):
        col = row.find_all("td")
        year = col[0].text
        revenue = col[1].text
        revenue_data.append({"Year": year, "Revenue": revenue})

    # Create DataFrame
    revenue_df = pd.DataFrame(revenue_data)

    # Data cleaning
    revenue_df["Revenue"] = revenue_df["Revenue"].str.replace(',|\$',"")
    revenue_df.dropna(inplace=True)
    revenue_df = revenue_df[revenue_df["Revenue"] != ""]

    return revenue_df

# URL for GameStop revenue data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Call the function to fetch GameStop revenue data
gamestop_revenue = gme_revenue(url)

# Display the last five rows of the DataFrame
print(gamestop_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Close Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    
    # Customize x-axis to show from 2010
    plt.xlim('2010-01-01')
    
    plt.show()

tsla = yf.Ticker('TSLA')   # Fetch Tesla data
tsla_data = tsla.history(start='2010-01-01')  # Fetch Tesla data starting from 2010

# Plot the graph
make_graph(tsla_data, title='Tesla Stock Price')
```
